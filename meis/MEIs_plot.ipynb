````python
import matplotlib.pyplot as plt

# Tên file đầu vào
file1_name = "vn_007.line1.txt"
file2_name = "vn_0007.xtea.short.line1.txt"
output_file_name = "intersection_ids.txt"

def read_file_to_set(file_path):
    """Đọc tệp và trả về một tập hợp (set) các dòng."""
    try:
        with open(file_path, 'r') as f:
            # Đọc từng dòng, loại bỏ khoảng trắng/ký tự xuống dòng ở đầu và cuối
            data_set = set(line.strip() for line in f if line.strip())
        return data_set
    except FileNotFoundError:
        print(f"Lỗi: Không tìm thấy tệp tại {file_path}")
        return set()

# Đọc dữ liệu vào các tập hợp
set1 = read_file_to_set(file1_name)
set2 = read_file_to_set(file2_name)

# Tính toán các phép toán tập hợp
intersection = set1.intersection(set2)
unique_to_file1 = set1 - set2
unique_to_file2 = set2 - set1

# --- Bước 1: Xuất danh sách ID giao thoa ra file ---
with open(output_file_name, 'w') as out_f:
    # Ghi từng ID giao thoa vào tệp, mỗi ID một dòng
    out_f.write('\n'.join(sorted(list(intersection))))

print(f"--- Đã lưu {len(intersection)} ID giao thoa vào file '{output_file_name}' ---")

# --- Bước 2: Chuẩn bị dữ liệu và tạo biểu đồ cột ---
# Dữ liệu cho Biểu đồ Cột (Các thành phần của Venn)
venn_components = {
    f'Chỉ có trong {file1_name}': len(unique_to_file1),
    'Giao thoa (Chung)': len(intersection),
    f'Chỉ có trong {file2_name}': len(unique_to_file2)
}

component_names = list(venn_components.keys())
component_values = list(venn_components.values())

# Tạo biểu đồ cột
plt.figure(figsize=(10, 6))
plt.bar(component_names, component_values, color=['skyblue', 'gold', 'lightcoral'])
plt.title('So sánh ID Duy nhất và ID Chung giữa hai Tập tin')
plt.ylabel('Số lượng ID Duy nhất')
plt.xticks(rotation=15, ha='right')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()

# Lưu biểu đồ
plt.savefig('venn_bar_comparison_final.png')

# In kết quả tổng hợp
print("\n--- Kết quả Phân tích Tập hợp ---")
print(f"Tổng số ID trong {file1_name}: {len(set1)}")
print(f"Tổng số ID trong {file2_name}: {len(set2)}")
print(f"Số lượng ID Giao thoa (Chung): {len(intersection)}")
print(f"Số lượng ID Chỉ có trong {file1_name}: {len(unique_to_file1)}")
print(f"Số lượng ID Chỉ có trong {file2_name}: {len(unique_to_file2)}")
print(f"Tổng số ID Hợp nhất (Union): {len(set1.union(set2))}")
print("\nĐã lưu biểu đồ cột vào file 'venn_bar_comparison_final.png'.")
````

````python
print("\nAnalyzing proportional distribution of 'Sub_Region' within each 'Type':")

# Group by 'Type' and 'Sub_Region' and get the 'Proportion'
grouped_proportions = df_clean.groupby(['Type', 'Sub_Region'])['Proportion'].sum()

for type_val in df_clean['Type'].unique():
    print(f"\n--- {type_val} ---")

    # Enriched regions (highest proportions)
    enriched_regions = grouped_proportions[type_val].sort_values(ascending=False)
    print(f"Most enriched 'Sub_Region's for {type_val}:\n{enriched_regions.head()}\n")

    # Depleted regions (lowest proportions)
    depleted_regions = grouped_proportions[type_val].sort_values(ascending=True)
    print(f"Most depleted 'Sub_Region's for {type_val}:\n{depleted_regions.head()}")
````

**Reasoning**:
The previous code failed with a `UserWarning` because `sns.histplot` in this specific configuration (categorical `y`, `weights`, `hue`, `multiple='stack'`) does not generate legend artists that `plt.legend()` can automatically pick up. To resolve this, I will adapt the successful `sns.barplot` loop approach from earlier cells to manually construct the stacked bar chart, ensuring each segment is labeled correctly for `plt.legend()`.



````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create a figure with a specified size
plt.figure(figsize=(12, 8))

# Determine the order for stacking. This is important for visual consistency.
region_order = sorted(df_clean['Sub_Region'].unique())

# Initialize the bottom for stacking for each 'Type'
# The index should be the unique 'Type' values from df_clean
type_order = sorted(df_clean['Type'].unique())
bottoms = pd.Series(0.0, index=type_order)

# Use a consistent color palette
palette = sns.color_palette("tab10", len(region_order))
color_map = {region: palette[i] for i, region in enumerate(region_order)}

# Iterate through each sub-region and plot its proportion segment
for region in region_order:
    # Get proportions for the current region, ensuring all types are present
    current_region_props = df_clean[df_clean['Sub_Region'] == region].set_index('Type')['Proportion'].reindex(type_order, fill_value=0)

    # Plot using sns.barplot (horizontal stacked bar)
    sns.barplot(
        y=current_region_props.index,  # 'Type' on the y-axis
        x=current_region_props.values, # Proportions on the x-axis
        color=color_map[region], # Consistent color for the region
        left=bottoms, # Stack on previous segments
        label=region, # Label for the legend
        ax=plt.gca() # Plot on the current active axes
    )

    # Update the bottom for the next region
    bottoms += current_region_props.values

# Customize the plot
plt.title('Proportional Distribution of Sub-Regions by Type')
plt.xlabel('Proportion')
plt.ylabel('Type (ALU, LINE1, SVA)')

# Set x-axis ticks to represent percentages (0.0 to 1.0)
plt.xticks([0.0, 0.25, 0.5, 0.75, 1.0], ['0%', '25%', '50%', '75%', '100%'])
plt.xlim(0, 1) # Ensure x-axis goes from 0 to 1 for proportions

# Move the legend outside the plot
plt.legend(title='Sub-Region', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout to prevent plot overlaps
plt.tight_layout()

# Display the plot
plt.show()
````
